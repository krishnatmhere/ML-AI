```python
#split data and label
data_val = data.drop('price_range',axis=1)
label = data['price_range']
```

```python
data_val.shape
```

```python
label.shape
```

```python
from keras.utils import to_categorical
# one hot encode output variable
y = to_categorical(label)
```

```python
#standardization
from sklearn.preprocessing import StandardScaler
scalar = StandardScaler()
data_val = scalar.fit_transform(data_val)
```

```python
data_val[:5,:]
```

```python
#split data for training and testing
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test = train_test_split(data_val,y,test_size=0.25,random_state=10)
```

```python
x_train.shape
```

```python
#package for creating model
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(10,input_dim=20,activation='relu'))
model.add(Dense(4,activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy',optimizer='adam',metrics=['accuracy'])
```

```python
history = model.fit(x_train,y_train,validation_split=0.2,epochs=200)
```

```python
y_pred = model.predict(x_test)
```

```python
y_pred
```

```python
y_train
```

```python
_, train_acc = model.evaluate(x_train, y_train, verbose=0)
_, test_acc = model.evaluate(x_test, y_test, verbose=0)
```

```python
print('Train: %.3f, Test: %.3f' % (train_acc, test_acc))
```

```python
from matplotlib import pyplot
# plot loss during training
pyplot.subplot(211)
pyplot.title('Loss')
pyplot.plot(history.history['loss'], label='train')
pyplot.plot(history.history['val_loss'], label='test')
pyplot.legend()
```

```python
# plot accuracy during training
pyplot.subplot(212)
pyplot.title('Accuracy')
pyplot.plot(history.history['accuracy'], label='train')
pyplot.plot(history.history['val_accuracy'], label='test')
pyplot.legend()
pyplot.show()
```
